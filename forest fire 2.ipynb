```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for column in a.columns:
    print("--------------------")
    print(f"{column} : {a[column].unique()}")
```

```python
# pie chart
print( a['Fire Occurrence'].value_counts())
```

```python
sizes=[20,19]
labels='Fire','No Fire'
plt.pie(sizes,labels=labels,colors=['red','yellow'],autopct="%1.1f%%")
```

```python
fire_counts = a.groupby('Area')['Fire Occurrence'].sum()

# Plot bar graph
plt.figure(figsize=(12, 6))
plt.bar(fire_counts.index, fire_counts.values)
plt.title('Fire Occurrence by Area')
plt.xlabel('Area')
plt.ylabel('Number of Fire Occurrences')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'a' is your DataFrame
plt.scatter(a['Temperature'], a['Humidity'])
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Scatter Plot of Temperature vs Humidity')
plt.grid(True)
plt.show()
```

```python
sns.set(style='ticks')
sns.heatmap(a.corr(),annot=True)
```

```python
numerical_columns = ['Oxygen', 'Temperature', 'Humidity']
box_data = a[numerical_columns]

# Plotting box plot using Seaborn
sns.boxplot(data=box_data)
plt.title('Box Plot of Oxygen, Temperature, and Humidity')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
sns.pairplot(a)
```

```python
mean_temp_by_area = a.groupby('Area')['Temperature'].mean()

# Plot line plot
plt.figure(figsize=(10, 6))
plt.plot(mean_temp_by_area.index, mean_temp_by_area.values, marker='o', linestyle='-')
plt.title('Mean Temperature by Area')
plt.xlabel('Area')
plt.ylabel('Mean Temperature')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
