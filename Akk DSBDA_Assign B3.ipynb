# **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

# **Reading Data**

```python
a = pd.read_csv(r"D:\Abhishek\airquality.csv")
a
```

# **Data Cleaning**

```python
a = a.drop(['Unnamed: 0'],axis=1)
a
```

```python
a['Ozone'] = a['Ozone'].fillna(a['Ozone'].mean())
```

```python
a['Solar.R'] = a['Solar.R'].fillna(a['Solar.R'].mean())
```

```python
a['Humidity'] = a['Humidity'].fillna(a['Humidity'].mode()[0])
```

```python
a.isnull().sum()
```

```python
a
```

# **1) Pie Chart**

```python
labels = 'Ozone','Temperature','Wind','Solar.R'
colors = ('red','purple','green','yellow')
sizes = [a['Ozone'].mean(),a['Temp'].mean(),a['Wind'].mean(),a['Solar.R'].mean()]
explode = (0.1,0,0,0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 0)
plt.title('Average Data')
```

# **2) Bar Plot**

```python
b = a.iloc[:21,3]
y = np.arange(len(b))
v = range(0,21)
```

```python
plt.bar(y,b,align = 'center',alpha=0.5)
plt.xticks(y,v)
plt.ylabel('Temperature')
plt.xlabel('Days')
plt.title('Temperature for 1st-20th May ')
```

# **3) Histogram**

```python
c = a.iloc[:,0]
plt.hist(c,bins = 'auto')
plt.xlabel('Ozone')
plt.ylabel('No. of values')
plt.title('Histogram for Ozone values')
```

# **4) Line Graph**

```python
l = a.iloc[:20,0]
m = a.iloc[:20,2]
plt.plot(l, label = 'Ozone', marker = '*', linestyle = 'dashdot')
plt.plot(m, label = 'Wind', marker = 'o', linestyle = 'solid')
plt.legend()
plt.title('Ozone and wind chart for 1st-20th May')
plt.ylabel('Days')
plt.xlabel('Values')
```

# **5) Scatter Plot**

```python
p = a.iloc[:,0]
t = a.iloc[:,3]
plt.scatter(p,t)
plt.title('Ozone vs Temperature')
plt.xlabel('Ozone')
plt.ylabel('Temperature')
```

# **6) Box Plot**

```python
a.boxplot(by = 'Month', column = ['Wind'], grid = False)
plt.xlabel('Month')
plt.ylabel('Wind values')
plt.title('Box plot for wind grouped by month')
```

# **7) Pair Plot**

```python
sb.set(style = 'ticks')
sb.pairplot(a)
```

# **8) Heat Map**

```python
corr = a.corr()
sb.heatmap(corr, annot = True)
```
